# Customization

1. Built-in Styles
Seaborn has five preset themes:

darkgrid (Default): Grey background with white grid lines.

whitegrid: White background with grey grid lines (great for plots with lots of data).

dark: Solid grey background without lines.

white: Solid white background without lines.

ticks: Similar to white, but adds small marks on the axes.

```python
# Fill in the style name here
sns.set_style("whitegrid")
sns.set_context("notebook")
```

Digging into Qualitative Palettes
Since these are used for categories with no specific order, they use distinct colors to make groups easy to tell apart. Some popular built-in qualitative palettes include:

deep

muted

bright

pastel

colorblind (Highly recommended for accessibility!)

```python
sns.set_palette("deep")
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Creating mock data
data = {
    'Region': ['North', 'South', 'East', 'West'],
    'Sales': [450, 600, 300, 550]
}
df = pd.DataFrame(data)
```

```python
sns.barplot(data=df, x='Region', y='Sales', palette='pastel')
plt.show()
```

Now, let's look at a different scenario. Imagine our dataset changed from simple regions to Customer Satisfaction Scores ranging from 1 to 10. Since 10 is much "higher" or "better" than 1, we want the colors to reflect that intensity.

This is where a Sequential palette comes in. These palettes use a single color that gets darker or more intense as the numbers increase.

Common sequential palettes in Seaborn include:

"Blues"

"Greens"

"Reds"

"YlOrBr" (Yellow to Orange to Brown)

If we wanted our bars to go from a very light blue for low sales to a deep, dark blue for high sales, which palette name do you think we should use?

```python
# Sequential example
sns.barplot(data=df, x='Region', y='Sales', palette='Reds')
```

3. Diverging Palettes
Now, let's look at the third type. Imagine we add a new column to our mock data called "Profit_Margin". Some regions made a profit (positive numbers), and some had a loss (negative numbers).

We want the "neutral" point (zero) to be a light color, while the extreme ends (big profit vs. big loss) are two different dark colors. This is a Diverging palette. 📉

Common diverging palettes include:

"RdBu" (Red to Blue)

"PRGn" (Purple to Green)

"coolwarm"

```python
data = {
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun'],
    'Profit_Margin': [-15, 10, -5, 25, 30, -10]
}
df_profit = pd.DataFrame(data)
```

```python
sns.barplot(data=df_profit, x='Month', y='Profit_Margin', hue='Profit_Margin', palette='RdBu')
```
