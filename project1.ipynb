# Importing data

```python
import pandas as pd 
import numpy as np 


# load the dataset 

review_df = pd.read_csv("datasets/new-york-city-airbnb-market/airbnb_last_review.csv")
price_df = pd.read_csv("datasets/new-york-city-airbnb-market/airbnb_price.csv")
room_df = pd.read_excel("datasets/new-york-city-airbnb-market/airbnb_room_type.xlsx")

print('Review dataset shape:', review_df.shape)
print('Price dataset shape:', price_df.shape)
print('Room dataset shape:', room_df.shape)
```

```python
# processing the review dataset 

# 0. stripping the leading and trailing spaces from the columns names 
review_df.columns = review_df.columns.str.strip(' ')

# 1. convert the date column to datetime format
review_df['last_review'] = pd.to_datetime(review_df['last_review'] , errors = 'coerce').dt.date

# 2. changing the dtype of listings_id to string
review_df['listing_id'] = review_df['listing_id'].astype(str)

# 3. checking for missing values 
missing_values = review_df.isnull().sum()

# 4. checking duplicated values 
duplicated_values = review_df.duplicated().sum()

print(review_df.head())
print("Missing values in each column:")
print(missing_values)
print("Duplicated values:")
print(duplicated_values)
```

```python
# processing the review dataset 

# 0. stripping the leading and trailing spaces from the columns names 
price_df.columns = price_df.columns.str.strip(' ')

# 1. removing the $ sign from the price column and converting it to numeric
price_df['price'] = price_df['price'].str.strip('dollars').astype(float)

# 2. changing the dtype of listings_id to string
price_df['listing_id'] = price_df['listing_id'].astype(str)

# 3. checking for missing values 
missing_values = price_df.isnull().sum()

# 4. checking duplicated values 
duplicated_values = price_df.duplicated().sum()
print(price_df.head())
print("Missing values in each column:")
print(missing_values)
print("Duplicated values:")
print(duplicated_values)
```

```python
# 1. converting the description column to string type 
room_df['description'] = room_df['description'].astype(str)

# 2. converting the listing_id to string type 
room_df['listing_id'] = room_df['listing_id'].astype(str)

# 3. coverting the room_type column into categorical type 
room_df['room_type'] = room_df['room_type'].astype('category')

# 4. checking for missing values 
missing_values = room_df.isnull().sum()
# 5. checking for duplicated values
duplicated_values = room_df.duplicated().sum()

print(room_df.head())
print("Missing values in each column:")
print(missing_values)
print("Duplicated values:")
print(duplicated_values)
```

# Combining data

```python
df_all_concat = pd.concat([review_df , price_df , room_df] , axis = 1)

print(df_all_concat.head())
print(len(df_all_concat.columns))
```

| how     | meaning            |
| ------- | ------------------ |
| `inner` | only matching rows |
| `left`  | keep all from df1  |
| `right` | keep all from df2  |
| `outer` | keep everything    |


```python
price_df
```

```python
df_all_merge = review_df.merge(price_df , on = 'listing_id', how = 'inner').merge(room_df , on = 'listing_id' , how = 'inner')

print(df_all_merge.head())
print(len(df_all_merge.columns))
print('Shape of merged dataset:', df_all_merge.shape)
```
